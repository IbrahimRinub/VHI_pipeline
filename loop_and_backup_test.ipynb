```python
# copy files from one destination to other destination

import shutil
import os
from datetime import datetime



now = datetime.now()
dt_string = now.strftime("%d.%m.%Y_%H.%M") # .%S
# path to source directory
src_dir = r'C:\Users\Asus TUF\Desktop\DA'
# path to destination directory
dest_dir = r'C:\Users\Asus TUF\Desktop\DA\Archive'

os.chdir(src_dir)

for f in os.listdir():                
    file_name, file_exc = os.path.splitext(f)
    size = len(file_name)
    
    if size == 15 or size == 10:
        try:
            source = src_dir + '\\' + file_name + file_exc
            file_name = file_name + '_'+ dt_string + file_exc
            destination = dest_dir + '\\' + file_name
            print(destination)
        except:
            pass
        finally:
            shutil.copy(source, destination)
            print('copyed successfully')
        
                
```
